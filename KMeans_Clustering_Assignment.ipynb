# KMeans Clustering Assignment

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.cluster import KMeans
```

```python
df = pd.read_csv('customers.csv')
```

```python
df.head()
```

```python
null_counts = df.isnull().sum()
```

```python
print("Null values in each column:")
print(null_counts)
```

```python
df = df.dropna()
```

```python
plt.scatter(df['Age'], df['Spending Score (1-100)'])
plt.xlabel('Age')
plt.ylabel('Spending Score (1-100)')
plt.title('Age vs Spending Score')
plt.show()
```

```python
X = df[['Age', 'Spending Score (1-100)']]

kmeans = KMeans(n_clusters=5)  # Specify the number of clusters
kmeans.fit(X)
df['Cluster'] = kmeans.labels_

plt.scatter(df['Age'], df['Spending Score (1-100)'], c=df['Cluster'], cmap='viridis')
plt.xlabel('Age')
plt.ylabel('Spending Score (1-100)')
plt.title('Age vs Spending Score (Clustered)')
plt.colorbar(label='Cluster')
plt.show()
```

```python
plt.scatter(df['Age'], df['Spending Score (1-100)'])
plt.xlabel('Age')
plt.ylabel('Spending Score (1-100)')
plt.title('Age vs Spending Score')
plt.show()

# Prepare data for clustering
X = df[['Age', 'Spending Score (1-100)']]

# Find the best number of clusters using the elbow method
inertia = []
for k in range(1, 11):
    kmeans = KMeans(n_clusters=k)
    kmeans.fit(X)
    inertia.append(kmeans.inertia_)

# Plot the elbow curve
plt.plot(range(1, 11), inertia, marker='o')
plt.xlabel('Number of Clusters')
plt.ylabel('Inertia')
plt.title('Elbow Method for Optimal K')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.cluster import AgglomerativeClustering
from scipy.cluster.hierarchy import dendrogram, linkage

iris = load_iris()
X = iris.data 

model = AgglomerativeClustering(n_clusters=3)  
model.fit(X)

linked = linkage(X, 'ward')  
plt.figure(figsize=(10, 7))
dendrogram(linked, orientation='top', labels=iris.target, distance_sort='descending', show_leaf_counts=True)
plt.title('Dendrogram for Iris Dataset')
plt.xlabel('Species')
plt.ylabel('Distance')
plt.show()
```

# Dendrogram Assignment

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.cluster.hierarchy import dendrogram, linkage

# Data
data = np.array([[5, 3],
                 [10, 15],
                 [15, 12],
                 [24, 10],
                 [30, 30],
                 [85, 70],
                 [71, 80],
                 [60, 78],
                 [70, 55],
                 [80, 91]])

plt.figure(figsize=(8, 6))
plt.scatter(data[:, 0], data[:, 1], s=100)
plt.title('Scatter Plot of Data')
plt.xlabel('X')
plt.ylabel('Y')
plt.grid(True)
plt.show()

linked = linkage(data, 'ward') 
plt.figure(figsize=(10, 7))
dendrogram(linked, orientation='top', distance_sort='descending', show_leaf_counts=True)
plt.title('Dendrogram for Agglomerative Clustering')
plt.xlabel('Data Points')
plt.ylabel('Distance')
plt.show()
```

# Case Study 2 - K-means

```python
import pandas as pd
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

mtcars = pd.read_csv('cars-1 (1).csv')

car_features = mtcars[['mpg', 'disp', 'hp']]

kmeans_3 = KMeans(n_clusters=3)
kmeans_3.fit(car_features)
clusters_3 = kmeans_3.predict(car_features)

car_features['cluster_3'] = clusters_3

cluster_0 = car_features[car_features['cluster_3'] == 0]
cluster_1 = car_features[car_features['cluster_3'] == 1]
cluster_2 = car_features[car_features['cluster_3'] == 2]

plt.scatter(cluster_0['mpg'], cluster_0['disp'], color='red', label='Cluster 0')
plt.scatter(cluster_1['mpg'], cluster_1['disp'], color='blue', label='Cluster 1')
plt.scatter(cluster_2['mpg'], cluster_2['disp'], color='green', label='Cluster 2')
plt.xlabel('mpg')
plt.ylabel('disp')
plt.title('K-means Clustering with 3 clusters')
plt.legend()
plt.show()

kmeans_5 = KMeans(n_clusters=5)
kmeans_5.fit(car_features)
clusters_5 = kmeans_5.predict(car_features)

car_features['cluster_5'] = clusters_5

for i in range(5):
    cluster_i = car_features[car_features['cluster_5'] == i]
    print(f"Observations in Cluster {i}:", len(cluster_i))
```

```python
import pandas as pd
from surprise import Dataset, Reader, SVD
from surprise.model_selection import cross_validate
from copy import deepcopy

# Load ratings and movies datasets
ratings = pd.read_csv('ratings.csv')
movies = pd.read_csv('movies.csv')

# Find unique number of users and movies
num_users = ratings['userId'].nunique()
num_movies = ratings['movieId'].nunique()

# Create rating matrix
Ratings = ratings.pivot(index='userId', columns='movieId', values='rating').fillna(0)

# Load ratings dataset as SVD's Dataset object
reader = Reader(rating_scale=(0.5, 5))
data = Dataset.load_from_df(ratings[['userId', 'movieId', 'rating']], reader)

# Compute 3-fold cross-validation using SVD object
algo = SVD()
cross_validate(algo, data, measures=['RMSE', 'MAE'], cv=3, verbose=True)

# Find movies rated as 5 stars by user id 5
ratings_1 = ratings[(ratings['userId'] == 5) & (ratings['rating'] == 5)]

# Create a shallow copy of movies dataset for user 5
user_5 = deepcopy(movies)

# Train recommender system using SVD object
trainset = data.build_full_trainset()
algo.fit(trainset)

# Predict ratings for user id 5
user_5['predicted_rating'] = user_5['movieId'].apply(lambda x: algo.predict(5, x).est)

# Print top 10 movie recommendations for user id 5
top_10_recommendations = user_5.sort_values(by='predicted_rating', ascending=False).head(10)
print(top_10_recommendations)
```
